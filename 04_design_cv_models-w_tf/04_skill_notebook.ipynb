```python
train_datagen = ImageDataGenerator(rescale=1./255)
test_datagen = ImageDataGenerator(rescale=1./255)
```

```python
train_generator = train_datagen.flow_from_directory(
    'ramen_sushi/train/',
    target_size=(200, 200),
    batch_size=32,
    class_mode='binary'
)
```

```python
test_generator = test_datagen.flow_from_directory(
    'ramen_sushi/test/',
    target_size=(200, 200),
    batch_size=32,
    class_mode='binary'
)
```

```python
print(f"Total train images: {train_generator.samples}")
print(f"Total test images: {test_generator.samples}")
print(f"Classes: {train_generator.class_indices}")
```

```python
def plot_images(images, labels, classes, num_images=5):
    plt.figure(figsize=(10, 10))

    for i in range(num_images):
        plt.subplot(1, num_images, i+1)
        plt.imshow(images[i])

        label_index = int(labels[i])
        plt.title(classes[label_index])
        plt.axis('off')
    plt.show()
```

```python
sample_training_images, sample_training_labels = next(train_generator)

class_names = ['ramen', 'sushi']
```

```python
plot_images(sample_training_images, sample_training_labels, class_names)
```
